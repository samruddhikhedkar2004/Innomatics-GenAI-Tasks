# Task 1: Conceptual Understanding

### **1. Difference between "Love" and "love"**

In NLP, "Love" and "love" are treated as different words because of the difference in capitalization. For example, "Love" might appear at the beginning of a sentence or be used as a proper noun, while "love" is usually used as a regular word. If we don't convert text to lowercase during preprocessing, the model treat them as separate words, which can increase the number of unique words unnecessarily and affect the model’s performance.

That's why in most NLP tasks, we convert all text to lowercase so that words like "Love" and "love" are treated the same.

### **2. What happens if stopwords are not removed?**

If you don't remove stopwords, words like "is", "the", "I", "a" stay in your data. These words appear everywhere but carry no real meaning. Because of this the model get confused by noise, which can reduce efficiency and sometimes affect performance.

So in short not removing stopwords doesn't break your code, but it hurts your model's quality


### **3. Provide two real-world scenarios where removing stopwords can be harmful.**

i. Search Engines / Question Answering:

Imagine someone searches "To be or not to be". If stopwords are removed, "to", "be", "or", "not" all get deleted and you're left with almost nothing meaningful. The search completely fails to find what the user actually wanted.

Same thing with a question like "How to do this?" after stopword removal it becomes just "do", which is basically useless.

ii. Machine Translation:

When translating between languages, sentence structure and grammar matter a lot. Words like "is", "the", "a", "not" are technically stopwords but they carry grammatical meaning.

For example: "He is not happy" vs "He is happy", the word "not" is often listed as a stopword, but removing it completely flips the meaning. In translation tasks, that's a serious problem

### **4. Explain the difference between stemming and lemmatization**

Stemming and lemmatization are both techniques used to reduce words to their base form, but they work in different ways.

Stemming is a simple and fast process that removes the ending of a word without considering its meaning. For example, words like "running", "runner", and "runs" all get reduced to "run" or sometimes even an incomplete form like "runn".

Lemmatization, on the other hand, is more advanced and considers the meaning and context of the word. It converts words into their correct base or dictionary form. For example, "better" becomes "good" and "running" becomes "run".

In short, stemming is faster but less accurate, while lemmatization is slower but gives more meaningful results.

# Task 2: Build Advanced Preprocessing Function

```python
def preprocess_text(text):
    # Handle empty or invalid input
    if not text or not isinstance(text, str):
        return [], ""
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Remove special characters (including emojis)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 8. Tokenization
    tokens = text.split()
    
    # 9. Remove short words (keep 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    # 10. Join back to sentence
    cleaned_sentence = " ".join(tokens)
    
    return tokens, cleaned_sentence
```

# Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("Stress Testing Results:\n")

count = 1

for sentence in test_sentences:
    
    tokens, cleaned = preprocess_text(sentence)
    
    print("Example", count)
    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-----------------------------------")
    
    count += 1
```

# Task 4: Token Analytics

```python
print("\nToken Analytics:\n")

for sentence in test_sentences:
    
    tokens, cleaned = preprocess_text(sentence)
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    # calculating average token length
    if total_tokens > 0:
        total_length = 0
        for word in tokens:
            total_length += len(word)
        avg_length = total_length / total_tokens
    else:
        avg_length = 0
    
    print("Sentence:", sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-----------------------------------")
```

# Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

# collecting tokens from all sentences
for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    all_tokens.extend(tokens)

# counting frequency
freq = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
# converting to list and taking last 5
least_common = freq.most_common()[-5:]
print(least_common)
```

# Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        
        tokens, cleaned = preprocess_text(text)
        
        # storing all tokens together
        all_tokens.extend(tokens)
        
        # storing cleaned sentences
        clean_sentences.append(cleaned)
    
    result = {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
    
    return result
```

```python
output = full_pipeline(test_sentences)

print("All Tokens:")
print(output["tokens"])

print("\nClean Sentences:")
print(output["clean_sentences"])
```
